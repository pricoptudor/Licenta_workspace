```python
import spotipy
import requests
from spotipy.oauth2 import SpotifyClientCredentials
from mutagen.mp3 import MP3
from mutagen.id3 import ID3, TIT2, TPE1

# Set up client credentials
client_id = "acb72e83d144464f910c84091169c667"
client_secret = "1368e8bdbedf4b81b2eff3a41c3b5876"
client_credentials_manager = SpotifyClientCredentials(client_id, client_secret)
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)

genres = ['blues', 'classical', 'country', 'disco', 'reggae', 'metal',
           'hip-hop', 'jazz', 'pop', 'rock']

AUDIO_PATH = '../Proiect_3/Data/Audio_Spotify_Test/'

for genre in genres:
    audio_counter = 0
    file_path = AUDIO_PATH + '/' + genre

    for offset in range(500, 800, 50):
        results = sp.search(q=f"genre:{genre}", type="track", limit=50, offset=offset)
        
        for track in results["tracks"]["items"]:
            track_id = track["id"]
            track_name = track["name"]
            artist_name = track["artists"][0]["name"]
            preview_url = track["preview_url"]

            if preview_url:
                audio_counter += 1
                filename = file_path + '/' + str(audio_counter) + '.mp3'
                with open(filename, "wb") as f:
                    f.write(requests.get(preview_url).content)
                    print(f"Downloaded {track_name} - {artist_name}.mp3")
                audio_file = MP3(filename, ID3=ID3)
                audio_file["TIT2"] = TIT2(encoding=3, text=track_name)
                audio_file["TPE1"] = TPE1(encoding=3, text=artist_name)
                audio_file.save()
                print(f"Added metadata to {filename}")
            else:
                print(f"No preview available for {track_name} - {artist_name}")
```
